# Grocery Store

```python
# Import Libraries
import tkinter
from tkinter import *
import random
from tkinter import messagebox

#  Set the size and Colour
base  =Tk()
base.title('Billing Software')
base.maxsize(width = 1000,height = 800)
base.minsize(width = 1000,height = 800)
bg = '#F0FFFF'

global l
l = []  
```

# Variables

```python


custom_name =StringVar()

custom_phone = StringVar()

bill_no = StringVar()
x = random.randint(1000,9999)
bill_no.set(str(x))
Value = StringVar()
Price = IntVar()
Quantity = IntVar()
 
```

# Functions

```python
    

def welcome():
    textarea.delete(1.0,END)
    textarea.insert(END,"                Welcome to Nandanwan Store")
    textarea.insert(END,f'\n\nBill No.: {bill_no.get()}',)
    textarea.insert(END,f'\nCustomer name: {custom_name.get()}')
    textarea.insert(END,f'\nPhone Number:{custom_phone.get()}')
    textarea.insert(END,'\n*********************************************************')
    textarea.insert(END,'\nProduct\t\tQuantity\t\tPrice')
    textarea.insert(END,'\n*********************************************************\n\n')
    textarea.configure(font = ('arial 15 bold',15))
    
    
def add_item():
    p = Price.get()
    q = Quantity.get()
    total = int(p) * int(q)
    l.append(total)
    if value_text.get() == '':
        messagebox.showerror('Please enter Product')
    else:
        textarea.insert((10.0+float(len(l)-1)),f'{value_text.get()}\t\t{quantity_text.get()}\t\t{total}\n')

                        
def get_bill():
    if custom_name.get() == '' or custom_phone.get() == '':
        messagebox.showerror('Please enter customer details')
    else:    
        txt = textarea.get(10.0,(10.0+float(len(l))))
        welcome()
        textarea.insert(END,txt)
        textarea.insert(END,'\n*********************************************************\n')
        textarea.insert(END,f'Total amount:        {sum(l)}')                  
        textarea.insert(END,'\n*********************************************************')
#         savebill()
        
# def savebill():
#     msg = messagebox.askyesno('Do you want to save the bill')
#     if msg>0:
#         details = textarea.get(1.0,END)
#         z = open("C:\Users\vaibh\Downloads\vaibhav_python_softronix\Project\Bills/"+str({bill_no.get()}))
#         z.write(details)
#         z.close()
#         messagebox.showinfo('Saved')
#     else:
#         return

    
def clear():
    custom_name.set('')
    custom_phone.set('')
    Value.set('')
    Price.set(0)
    Quantity.set(0)
    welcome()
    
```

# Labels and Buttons

```python

title = Label(base, text ='Billing Software',bg = bg,fg='black',font =('BankGothic Md BT',30),relief = GROOVE,bd = 5).pack(fill = X)
sub_frame = LabelFrame(base,text = "Customer Details",bg=bg,fg='black',font = ('Century Gothic',14),relief = GROOVE,bd = 5)
sub_frame.place(x = 0,y=80,relwidth = 1)


cname = Label(sub_frame,text = 'Customer Name',bg=bg,fg='black',font = ('Century Gothic',10))
cname.grid(row=0,column=0,padx = 30,pady=30)

cname_text = Entry(sub_frame,width = 20,font = ('Century Gothic',10),relief = SUNKEN,textvariable = custom_name)
cname_text.grid(row=0,column=1,padx = 30,pady=30)

cmob = Label(sub_frame,text = 'Mobile number',bg=bg,fg='black',font = ('Century Gothic',10))
cmob.grid(row=1,column=0)

cmob_text = Entry(sub_frame,width = 20,font = ('Century Gothic',10),relief = SUNKEN,textvariable = custom_phone)
cmob_text.grid(row = 1,column = 1)

sub_frame_2 = LabelFrame(base,text = "Product Details",bg=bg,fg='black',font = ('Century Gothic',14),relief = GROOVE,bd = 5)
sub_frame_2.place(x = 0,y=250,width = 550,height = 500)

value = Label(sub_frame_2,text ='Product Name',bg=bg,fg='black',font = ('Century Gothic',10))
value.grid(row=0,column=0,padx = 20,pady=20)

value_text = Entry(sub_frame_2,width = 20,font = ('Century Gothic',10),relief = SUNKEN,textvariable = Value)
value_text.grid(row = 0,column = 1,padx = 20,pady=20)


price = Label(sub_frame_2,text ='Price',bg=bg,fg='black',font = ('Century Gothic',10))
price.grid(row=1,column=0,padx = 20,pady=20)

price_text = Entry(sub_frame_2,width = 20,font = ('Century Gothic',10),relief = SUNKEN,textvariable = Price)
price_text.grid(row = 1,column = 1,padx = 20,pady=20)

quantity = Label(sub_frame_2,text ='Quantity',bg=bg,fg='black',font = ('Century Gothic',10))
quantity.grid(row=2,column=0,padx = 20,pady=20)

quantity_text = Entry(sub_frame_2,width = 20,font = ('Century Gothic',10),relief = SUNKEN,textvariable = Quantity)
quantity_text.grid(row = 2,column = 1,padx = 20,pady=20)


bt1 =Button(sub_frame_2,text ='Add item',font =('Century Gothic',10),command = add_item)
bt1.grid(row=3,column=0)

bt2 =Button(sub_frame_2,text ='Remove item',font =('Century Gothic',10),command = clear )
bt2.grid(row=3,column=1)

bt3 =Button(sub_frame_2,text ='Check Out',font =('Century Gothic',10),command = get_bill)
bt3.grid(row=3,column=2)

sub_frame_3 = Frame(base, relief = GROOVE,bd=10)
sub_frame_3.place(x=500,y=210,height=550,width = 500)

bill_title = Label(sub_frame_3,text =' Bill',font =('Century Gothic',15)).pack(fill = X)
scrol_y =Scrollbar(sub_frame_3,orient = VERTICAL)
textarea = Text(sub_frame_3,yscrollcommand = scrol_y)
scrol_y.pack(side = RIGHT, fill = Y)
scrol_y.config(command = textarea.yview)
textarea.pack()
welcome()

base.mainloop()
```
